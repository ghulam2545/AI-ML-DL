```python
import pydicom
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
images_path = '/content/drive/MyDrive/kaggle/test_images'
```

```python
dicom_files = tf.io.gfile.glob(images_path + '/*.dcm')
```

```python
len(dicom_files)
```

```python
sample = dicom_files[:10]

type(sample)
```

```python
one_image = pydicom.read_file(sample[0])

type(one_image)
```

```python
one_image
```

```python
print(one_image['PatientID'])
print(one_image['PatientSex'])
print(one_image['PatientAge'])
```

```python
plt.figure(figsize=(12, 5))
for i in range(len(sample)):
  curr_image = pydicom.read_file(sample[i])
  curr_age = curr_image.PatientAge
  curr_sex = curr_image.PatientSex
  plt.subplot(2, 5, i+1)
  plt.imshow(curr_image.pixel_array, cmap='bone')
  plt.title('Age: ' + curr_age + ', Sex: ' + curr_sex)
  plt.axis('off')
plt.show()  
```
